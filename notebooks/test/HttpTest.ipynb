```julia
# HTTP
using HTTP

HTTP.open("GET", "http://imagenet.stanford.edu/api/text/imagenet.synset.geturls?wnid=n02346627") do http
    for line in eachline(http)
        println(line)
    end
end
```

```julia
using PyCall
using PyPlot
plt = PyPlot.matplotlib.pyplot
io = pyimport("io")
Image = pyimport("PIL").Image

r = HTTP.request("GET", "https://farm1.static.flickr.com/201/496754650_999d27390c.jpg")
println(r.status)
img = Image.open(io.BytesIO(r.body))
plt.imshow(img)
```

```julia
url = "https://farm1.static.flickr.com/201/496754650_999d27390c.jpg"
r = HTTP.request("GET", url)
open(basename(url), "w") do fp
    write(fp, r.body)
end
```
